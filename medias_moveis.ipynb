```python
import yfinance as yf
import pandas as pd
```

```python
inicio = '2020-01-01'
fim = '2021-11-05'
```

```python
vale = yf.download('VALE3.SA', start=inicio, end=fim)
```

```python
vale.head()
```

```python
sma20 = vale['Adj Close'].rolling(20).mean()
sma60 = vale['Adj Close'].rolling(60).mean()
```

```python
vale['MM20'] = sma20
vale['MM60'] = sma60
```

```python
vale[['Adj Close','MM20','MM60']].plot(figsize=(10,10));
```
